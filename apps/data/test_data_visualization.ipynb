```python
import numpy as np

# Add a new column 'Side' based on conditions
df['sideWin'] = np.where((df['win'] & (df['teamId'] == 100)) | (~df['win'] & (df['teamId'] == 200)), 'blueSide', 'redSide')

# Print the DataFrame with the new column
df
```

```python
import pandas as pd

ROLES_PICK_COLUMNS = [
    "bluePickTop", "bluePickJungle", "bluePickMiddle", "bluePickBottom", "bluePickSupport", "redPickTop",
    "redPickJungle", "redPickMiddle", "redPickBottom", "redPickSupport"
]

games_played = df.shape[0]

# Create a Series of sets where each set represents unique bans in a row
unique_champions_picks_per_row = df.apply(lambda row: set(row[ROLES_PICK_COLUMNS]), axis=1)

# Concatenate all sets into a single Series
all_champions_picks = pd.Series([champion for row in unique_champions_picks_per_row for champion in row])

# Calculate cumulative sum for each Champion
cumulative_sum_champions_picks = all_champions_picks.value_counts()

# Create DataFrame with GamesBanned and Banrate
df_picks = pd.DataFrame({"GamesPicked": cumulative_sum_champions_picks, "Pickrate": cumulative_sum_champions_picks / games_played})

# Reset the index to make "ChampionName" a regular column
df_picks = df_picks.reset_index()

# Rename columns if needed
df_picks = df_picks.rename(columns={'index': 'ChampionName'})

# Print or use df_bans as needed
df_picks
```

```python
import pandas as pd

ROLES_BAN_COLUMNS = [
    'blueBanTop', 'blueBanJungle', 'blueBanMiddle', 'blueBanBottom', 'blueBanSupport', 'redBanTop', 'redBanJungle',
    'redBanMiddle', 'redBanBottom', 'redBanSupport'
]

games_played = df.shape[0]

# Create a Series of sets where each set represents unique bans in a row
unique_champions_bans_per_row = df.apply(lambda row: set(row[ROLES_BAN_COLUMNS]), axis=1)

# Concatenate all sets into a single Series
all_champions_bans = pd.Series([champion for row in unique_champions_bans_per_row for champion in row])

# Calculate cumulative sum for each Champion
cumulative_sum_champions_bans = all_champions_bans.value_counts()

# Create DataFrame with GamesBanned and Banrate
df_bans = pd.DataFrame({"GamesBanned": cumulative_sum_champions_bans, "Banrate": cumulative_sum_champions_bans / games_played})

# Reset the index to make "ChampionName" a regular column
df_bans = df_bans.reset_index()

df_bans = df_bans.rename(columns={'index': 'ChampionName'})

df_bans
```

```python
df_merged = pd.merge(df_picks, df_bans, on='ChampionName', how='left').fillna(0)

df_merged = df_merged.rename(columns={'GamesBanned': 'GamesBanned', 'Banrate': 'Banrate'})

df_merged["Presence"] = df_merged["Pickrate"] + df_merged["Banrate"]

df_merged.sort_values("Presence", ascending=False)
```

```python
import pandas as pd

red_side_values = df[df["sideWin"] == "redSide"]
red_side_values
red_columns = red_side_values.filter(like='redPick', axis=1)
red_columns
```

```python
df_melted_red = pd.melt(red_columns, value_vars=red_columns.columns)
red_side_champion_wins = df_melted_red['value'].value_counts()

red_side_champion_wins
```

```python
result = blue_side_champion_wins.add(red_side_champion_wins, fill_value=0).sort_values(ascending=False).astype(int)
result
```
