```python
import datetime

def convert_seconds_to_minutes_and_seconds(seconds: int) -> datetime.time:
    minutes = seconds // 60
    hours = minutes // 60
    remaining_seconds = seconds % 60

    return datetime.time(hour=hours, minute=minutes, second=remaining_seconds)
```
